**Data Overview**
The dataset is structured as follows
*   TransactionID - Indexing feature
*   isFraud - dependent variable ~3.5% fraud
*   Transaction Features:
    *   TransactionDT - Time since reference point
    *   TransactionAmt - Transaction Maount
    *   ProductCD - Product Category
*   Card and Account Features:
    * card1 – card6: Anonymized card-related attributes representing partial identifiers such as issuing bank, card type, and account-level characteristics. These can be used as proxy variables for user/account grouping but do not uniquely identify individuals.
    * addr1, addr2: Anonymized location-based features representing billing region or geographic information associated with the transaction.
*   Behavioral Features:
    *   C1 - C14 - number of past transactions
    *   D1 - D15 - time since last activity
*   V1 – V339 - Engineered Features
*   Device Features:
    *   DeviceType - Device used to make transaction
    *   DeviceInfo - Device OS information
    *   id_12 - id_38 - Device/transaction specific heuristics


**Transaction Information Aggregation**

* Event Features (Transaction-Level Signals):  
    * TransactionAmt, ProductCD  
    → Describe the characteristics of the transaction itself (size, type)

* Time Features (Temporal Context):  
    * TransactionDT  
    → Captures when the transaction occurred; used to model sequencing and behavioral patterns over time

* Identity / Account Proxies (User-Level Signals):  
    * card1–card6, addr1, addr2  
    * DeviceType, id_12–id_38  
    → Anonymized attributes used to approximate user identity, device fingerprinting, and account-level grouping

* Behavioral / Historical Features (Aggregated Activity Signals):  
    * C1–C14, D1–D15  
    → Precomputed features representing prior transaction behavior, frequency, and recency
